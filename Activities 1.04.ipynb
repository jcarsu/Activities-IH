1. Try to convert these functions into lambda functions

```python
def function_1(a,b):
    c=a+b
    return c
```

```python
a= 2
b= 7
function_1=lambda a,b : a+b
function_1(a,b)
```

```python
def function_2(a,b):
    c=0
    for i in range(a):
        c+=2*b
    return c
```

```python
function_2(a,b)
```

```python
function_2=lambda a,b: a*2*b 
function_2(a,b)
```

```python
def function_3(a,b):
    c=0
    if a>3:
        c=12
    else:
        c='Too big.'
    return c
```
