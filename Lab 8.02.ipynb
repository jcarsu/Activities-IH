### 5. List all language names and number of related articles in the order they appear in wikipedia.org:
    #url = 'https://www.wikipedia.org/'

```python
url = 'https://www.wikipedia.org/'
uscode = requests.get(url).content
soup = BeautifulSoup(uscode, "html.parser")
```

```python
#Sacamos lista de idiomas 
lang= soup.find_all('strong')
lang
```

```python
#Eliminamos primera y ultima fila
lang.pop(0)
lang.pop(10)
```

```python
#Creamos df de idiomas
lang=pd.DataFrame(lang)
lang
```

```python
#Llamamos los articulos
articles=soup.select('small > bdi')
articles
```

```python
#Creamos df de articulos
articles=pd.DataFrame(articles)
articles
```

```python
#Concatenamos ambas df
import numpy as np
wiki= np.concatenate((lang, articles),axis=1)
wiki
```

```python
#Creamos la df concatenada
wiki=pd.DataFrame(wiki)
wiki=wiki.set_axis(['Language', 'Articles'], axis=1)
wiki
```

### 7. Display the top 10 languages by number of native speakers stored in a pandas dataframe:
    #url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'

```python
#Extraemos la 'sopa' de la pagina web
url= 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
uscode= requests.get(url).content
soup= BeautifulSoup(uscode, "html.parser")
```

```python
#Llamamos la data
lang= soup.find_all('td')
lang
```

```python
#Creamos la lista
lang= [i.get_text() for i in lang]
lang
```

```python
lang=lang[0:108]
lang
```

```python
speakers= lang[1::4]
speakers
```

```python
lang=lang[0::4]
lang
```

```python
lang= pd.DataFrame({'Language': lang})
lang
```

```python
speakers= pd.DataFrame({'Native_speakers': speakers})
speakers
```

```python
langlist= np.concatenate((lang, speakers), axis=1)
langlist
```

```python
langlist= pd.DataFrame(langlist)
langlist= langlist.set_axis(['Language', 'Native_speakers'], axis=1)
langlist
```

```python
langlist= langlist.replace({'\n', ''}, regex=True)
langlist
```
