```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 1. 

```python
db=pd.read_csv('marketing_customer_analysis.csv')
db
```

```python
db.columns=db.columns.map(lambda x: x.lower())
db.dropna()
```

```python
db=db.drop(columns=['unnamed: 0','customer','vehicle type'],axis=1)
```

### 2.

```python
db.describe(include='all')
```

```python
db.isnull().sum()
```

```python
db.dtypes
```

```python
db['months since last claim']= db['months since last claim'].fillna(np.mean(db['months since last claim']))
db['number of open complaints']= db['number of open complaints'].fillna(np.mean(db['number of open complaints']))
```

```python
db['state'].fillna(db['state'].mode()[0], inplace=True)
db['response'].fillna(db['response'].mode()[0], inplace=True)
db['vehicle class'].fillna(db['vehicle class'].mode()[0], inplace=True)
db['vehicle size'].fillna(db['vehicle size'].mode()[0], inplace=True)
```

```python
db.isnull().sum()
```

```python
db['response'].hist(color='purple',bins=15,width=.1)
plt.xlabel('Answer')
plt.ylabel('Number of Responses')  
plt.title('Responses',
          fontweight ="bold")
plt.show()
```
