```python
def varianza (x):
  prom=sum(x)/len(x)
  #y-ŷ
  y_y=[]
  for i in x:
    y_y.append((i-prom)**2)
  return sum(y_y)/(len(x))
#jerarquía de operaciones matemáticas
#1- potencias y raíces
#2.- Mult y divisiones
#3.- sumas y restas
def SD (x):
  prom=sum(x)/len(x)
  #y-ŷ
  y_y=[]
  for i in x:
    y_y.append((i-prom)**2)
  return (sum(y_y)/(len(x)-1))**0.5
```

```python
lista=[159,160,200,150,198]
lista2=[182,182.2,182.3,182.1]
```

```python
#DEsvEst
varianza(lista)**0.5
```

```python
#DEsvEst
varianza(lista2)**0.5
```

```python
sns.heatmap(corr_matrix,annot=True,cmap='rainbow')
```

```python
maskp=np.zeros_like(corr_matrix)
maskp
```

```python
maskp[np.triu_indices_from(maskp)]=True
maskp
```
