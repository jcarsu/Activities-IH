```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('marketing_customer_analysis.csv')
df
```

```python
df.columns=[c.lower().replace(' ', '_') for c in df.columns]
df=df.drop(columns=['unnamed:_0', 'vehicle_type','customer'],axis=1)
df=df.dropna()
df
```

```python
df['effective_to_date']=pd.to_datetime(df['effective_to_date'], errors='coerce')
df['month']=df['effective_to_date'].dt.strftime('%m')
df
```

```python
df.describe(include='all')
```

```python
data_numerical= df.select_dtypes(exclude=["object"])
data_numerical
```

```python
data_numerical.isnull().sum()
```

```python
data_cat= df.select_dtypes(include=["object"])
data_cat
```

```python
data_cat.isnull().sum()
```

### 1.

```python
X=data_numerical.drop('total_claim_amount', axis=1)
y=data_numerical.total_claim_amount

X=X.drop(columns=['effective_to_date'])
X
```

### 2.

```python
X_num=X._get_numeric_data()
X_cat=X.drop(columns=X_num.columns)

print (X.shape, X_num.shape, X_cat.shape)  # check shape
```

```python
X_num.describe()
```

```python
# normalize (standard)
from sklearn.preprocessing import StandardScaler

transformer=StandardScaler().fit(X_num)
X_standardized=transformer.transform(X_num)

X_standardized_df=pd.DataFrame(X_standardized)
X_standardized_df
```

```python
(X_num-X_standardized).sum().sum()
```

### 3.

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder(drop='first').fit(X_cat)
encoder=encoder.transform(X_cat).toarray()
encoder
```

### 4.

```python
X=np.concatenate((X_standardized_df,encoder),axis=1)
X
```

## Linear Regression
### 1.

```python
from sklearn.model_selection import train_test_split as tts

X_train, X_test, y_train, y_test=tts(X, y, test_size=.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
len(X)
```

## 2.

```python
from sklearn.linear_model import LinearRegression as linreg
```

```python
model=linreg()
model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
```

## Model validation

### R2

```python
r_sq = model.score(X_test, y_test) # model prediction
r_sq
```

```python
r_sq = model.score(X_train, y_train) # model prediction
r_sq
```

```python
from sklearn.metrics import mean_squared_error as mse
MSE = mse(y_test,y_pred)
#MSE
#MSE = np.square(np.subtract(y_test,y_pred)).mean()
MSE
```
