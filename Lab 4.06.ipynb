```python
data_numerical=data.select_dtypes(exclude=object)
data_numerical
```

### 2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.

```python
sns.distplot(data_numerical['Customer Lifetime Value'], bins=5)
plt.show()

sns.distplot(data_numerical['Income'], bins=5)
plt.show()

sns.distplot(data_numerical['Monthly Premium Auto'], bins=5)
plt.show()

sns.distplot(data_numerical['Months Since Last Claim'], bins=5)
plt.show()

sns.distplot(data_numerical['Months Since Policy Inception'], bins=5)
plt.show()

sns.distplot(data_numerical['Number of Open Complaints'], bins=5)
plt.show()

sns.distplot(data_numerical['Number of Policies'], bins=5)
plt.show()

sns.distplot(data_numerical['Total Claim Amount'], bins=5)
plt.show()
```

```python
#All of the variables look more like an exponential distribution.
```

### 3. Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

```python
# Log-transform
def log_transform(x):
    if np.isfinite(x) and x!=0:
        return np.log(x)
    else:
        return np.NAN

for column in data_numerical:
    data_numerical[column] = data_numerical[column].apply(log_transform)
```

```python
data_numerical
```

```python
sns.distplot(data_numerical['Customer Lifetime Value'], bins=5)
plt.show()

sns.distplot(data_numerical['Income'], bins=5)
plt.show()

sns.distplot(data_numerical['Monthly Premium Auto'], bins=5)
plt.show()

sns.distplot(data_numerical['Months Since Last Claim'], bins=5)
plt.show()

sns.distplot(data_numerical['Months Since Policy Inception'], bins=5)
plt.show()

sns.distplot(data_numerical['Number of Open Complaints'], bins=5)
plt.show()

sns.distplot(data_numerical['Number of Policies'], bins=5)
plt.show()

sns.distplot(data_numerical['Total Claim Amount'], bins=5)
plt.show()
```

```python
#The distributions keep on looking more like exponential.
```
